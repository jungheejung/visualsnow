```python
import nibabel as nib
import numpy as np
from nilearn import plotting
# Load the corrected NIfTI image.
img = nib.load('/Users/heejungj/Documents/projects_local/visualsnow_source/derivatives/sub-002/anat/sub-002_desc-preproc_T1w.nii.gz')
data = img.get_fdata()

# Suppose you have the AC coordinate in voxel space (e.g., from manual identification).
# Replace these with the actual voxel coordinates of the AC.
ac_voxel = np.array([122, 149, 118])  # Example: [slice, row, column]

# Define crop size.
crop_dims = np.array([256, 256, 256])

# Compute starting indices so that the AC coordinate is centered.
start = ac_voxel - crop_dims // 2
end = start + crop_dims

# Make sure indices are within the bounds of your image.
start = np.maximum(start, 0)
end = np.minimum(end, data.shape)

# For a robust crop, you might need to adjust if the crop window extends beyond the image.
# Here we assume the image is large enough in each dimension.

# Crop the image.
# Note: Depending on your image orientation, the axis order might differ.
cropped_data = data[start[0]:end[0], start[1]:end[1], start[2]:end[2]]

# Save the cropped image.
cropped_img = nib.Nifti1Image(cropped_data, img.affine, header=img.header)
nib.save(cropped_img, 'HJcorrected_sub-002_desc-preproc_T1w.nii.gz')

print("Cropped image saved as 'cropped_image.nii'")
plotting.plot_anat('HJcorrected_sub-002_desc-preproc_T1w.nii.gz')
```
